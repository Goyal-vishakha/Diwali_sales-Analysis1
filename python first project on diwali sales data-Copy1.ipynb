# GENDER


```python
#to check how many members are female and how many are male using seaborn
```

```python
ax = sns.countplot(x='Gender',data=ds)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# to check how much amount spent by female and how much amount spent by male
```

```python
ds.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
sale_gen = ds.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sale_gen)
```

```python
# female spent more amount
```

# Age and Age_Group

```python
ds.columns
```

```python
# to check the count of age group
```

```python
ax= sns.countplot(x='Age Group',hue='Gender',data=ds)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# to check which age group spent more amount
```

```python
ds.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sale_age=ds.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',hue='Gender',data=sale_age)
```

```python
# so we have the age group 26-35 in which female have done more expenditure than males
```

# STATE have maore order and spent more money

```python
#total number of order according to the state
```

```python
sales_order=ds.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
plt.figure(figsize=(25,7))
sns.barplot(x='State',y='Orders',data=sales_order)
```

```python
total amount according to the state
```

```python
sales_order=ds.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(25,7))
sns.barplot(x='State',y='Amount',data=sales_order)
plt.show()
```

```python
# uttar pradesh, maharashtra ,Karnataka have done more expenditure  and order
```

# using Marital status check the amount

```python
sales_mari=ds.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(5,3))
sns.barplot(data=sales_mari,x='Marital_Status',y='Amount',hue='Gender')
plt.show()
```

# using occpation 

```python
ax=sns.countplot(x="Occupation",data=ds)
sns.set(rc={'figure.figsize':(35,10)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_occup=ds.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(35,10))
sns.barplot(data=sales_occup,x='Occupation',y='Amount')
plt.show()
```

```python
# IT sector,healthcare,aviation   has done more expennditure
```

# # Product Category

```python
ax=sns.countplot(x="Product_Category",data=ds)
sns.set(rc={'figure.figsize':(55,20)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_product=ds.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(35,10))
sns.barplot(data=sales_product,x='Product_Category',y='Amount')
plt.show()
```

```python
#people spent more on food  clothing and apparel 
```

```python
sales_product=ds.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(20)
plt.figure(figsize=(35,10))
sns.barplot(data=sales_product,x='Product_ID',y='Orders')
plt.show()
```
